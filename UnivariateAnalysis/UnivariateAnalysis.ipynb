Univariate Analysis - when we are doing a analysis of single column.

Bivariate Analysis - when we are doing a analysis of two column.

Multivariate Analysis - when we are doing a analysis of more than two column

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
df.head()
```

# 1.Categorical Data
a. Countplot

```python
# sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Embarked'])
```

b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
```
